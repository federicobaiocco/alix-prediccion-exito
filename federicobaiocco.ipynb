# Introducción

Les presentamos a la empresa “Frío Frío”, dedicada a la venta e instalación de equipos de aire acondicionado para grandes superficies. Al ser una empresa B2B (“Business To Business”), es esencial para ellos optimizar los esfuerzos de los representantes comerciales, ayudándolos a priorizar las oportunidades en el pipeline.

Una “oportunidad” consiste en un proyecto de venta o instalación de equipos para un cliente. La venta se estructura alrededor de TRF (Toneladas de refrigeración) y puede estar compuesta por varios productos distintos. El "pipeline" hace referencia al flujo de oportunidades prospecto que la empresa está desarrollando. El equipo comercial asigna a distintos momentos, para cada oportunidad, un estado en la negociación. En la Ilustración se muestran los estados que las oportunidades tienen dentro del pipeline.

![title](img/ilustracion1_alix_partners.png)

La variable que se está tratando de predecir es **“Probabilidad de éxito”** para cada oportunidad. ¿Cuál es la probabilidad de que la oportunidad se convierta en un caso Closed Won?
El dataset cuenta con información de cada oportunidad, como por ejemplo información sobre el vendedor a cargo de la venta, información geográfica de los clientes, TRS pedidas, fecha prevista de entrega de los equipos, etc. A partir de dichas variables es posible entrenar un modelo que prediga para un tiempo futuro el éxito o fracaso de cada oportunidad. Idealmente, “Frío Frío” podrá usar este modelo para predecir la probabilidad de éxito de cada oportunidad comercial, para mejorar el rendimiento y optimizar el esfuerzo de los vendedores.

El objetivo de esta competencia es, en base a los datos de oportunidades históricas, predecir el éxito de todas las oportunidade comerciales surgidas en los últimos cuatro meses.

# Helper functions

```python
def calc_prevalence(y):
    return (sum(y)/len(y))
```

```python
def miss_data(df, miss_value):
    x = ['variable','cantidad', 'porcentaje']
    missing_data = pd.DataFrame(columns=x)
    columns = df.columns
    for col in columns:
        icolumn_name = col
        imissing_data = df[df[col]==miss_value].shape[0]
        imissing_in_percentage = (df[df[col]==miss_value].shape[0]/df[col].shape[0])*100
        missing_data.loc[len(missing_data)] = [icolumn_name, imissing_data, imissing_in_percentage]
    return missing_data.set_index('variable').sort_values(by='porcentaje',ascending=False)
```

# Preprocesamiento

## Train preprocess pipeline

Definimos algunas clases para preprocesar el set de entrenamiento en un pipeline de sklearn:

```python
class UselessDropper( BaseEstimator, TransformerMixin ):
    '''
    Dropear filas que tienen stage != Closed Won y Closed Lost en el set de entrenamiento
    '''
    def __init__( self ):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        return X[X.Stage.isin(['Closed Won','Closed Lost'])]
```

```python
class OutliersHandler( BaseEstimator, TransformerMixin):
    '''
    Manejo de outliers en el set de entrenamiento
    '''
    def __init__( self ):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        X['ASP_(converted)'] = X['ASP_(converted)'].fillna(X['ASP_(converted)'].mean())
        X = X[X['ASP_(converted)'] != 0]
        X = X[X['ASP_(converted)'] < 1]
        return X
```

```python
class StageEncoder( BaseEstimator, TransformerMixin ):
    '''
    Encoding de la variable a predecir.
    '''
    def __init__( self):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        X['Stage'] = X['Stage'].replace({'Closed Won':1,'Closed Lost':0})
        return X
```

# Pipeline

Y las clases para armar el pipeline de preprocesamiento en general:

```python
class ColumnsDropper( BaseEstimator, TransformerMixin ):
    '''
    Dropear columnas que no se usan
    '''
    def __init__( self, features_to_drop ):
        self._features_to_drop = features_to_drop 
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        return X.drop(self._features_to_drop,axis=1 )
```

```python
class TotalTaxableAmountCalculator( BaseEstimator, TransformerMixin ):
    '''
    Porcentaje del total_amount que tiene impuestos.
    '''
    def __init__( self):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        X = X.dropna(subset=['Total_Amount','Total_Taxable_Amount'])
        X['Taxable_Amount_Percentage'] = X.Total_Amount / X.Total_Taxable_Amount
        X['Taxable_Amount_Percentage'] = X['Taxable_Amount_Percentage'].replace(np.inf, 0)
        X['Taxable_Amount_Percentage'] = X['Taxable_Amount_Percentage'].fillna(0)
        X = X.drop(['Total_Amount','Total_Taxable_Amount'], axis=1)
        return X
```

```python
class TimeElapsedCalculator( BaseEstimator, TransformerMixin ):
    '''
    Se calcula el tiempo transcurrido desde que se crea la oportunidad hasta que expira el presupuesto.
    '''
    def __init__( self):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        X['Quote_Expiry_Date'] = X['Quote_Expiry_Date'].fillna(X.Opportunity_Created_Date)
        X['created_to_expiry_days'] = X.Quote_Expiry_Date - X.Opportunity_Created_Date
        X['created_to_expiry_days'] = X['created_to_expiry_days'].dt.days.astype('int16')
        X = X[X['created_to_expiry_days'] >= 0]
        return X.drop(['Quote_Expiry_Date','Opportunity_Created_Date'], axis=1)
```

```python
class PlannedDeliveryDateProcessor( BaseEstimator, TransformerMixin ):
    '''
    Se obtienen distintas features de la fecha de entrega.
    '''
    def __init__( self):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        #X['Planned_Delivery_Month'] = X.Planned_Delivery_Start_Date.dt.month
        X['Planned_Delivery_Year'] = X.Planned_Delivery_Start_Date.dt.year
        #X['Planned_Delivery_Day_Of_Month'] = X.Planned_Delivery_Start_Date.dt.day
        #X['Planned_Delivery_Day_Of_Week'] = X.Planned_Delivery_Start_Date.dt.weekday
        X['Planned_Delivery_Quarter'] = X.Planned_Delivery_Start_Date.dt.quarter
        X = X.drop('Planned_Delivery_Start_Date', axis=1)
        return X
```

```python
class CreatedDateProcessor( BaseEstimator, TransformerMixin ):
    '''
    Se obtienen distintas features de la fecha de creación.
    '''
    def __init__( self):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        #X['Created_Month'] = X.Opportunity_Created_Date.dt.month
        X['Created_Year'] = X.Opportunity_Created_Date.dt.year
        X['Created_Quarter'] = X.Opportunity_Created_Date.dt.quarter
        X = X.drop('Opportunity_Created_Date', axis=1)
        return X
```

```python
class SalesContractTransformer( BaseEstimator, TransformerMixin ):
    '''
    Se rempazan los None por 0.
    Esta variable es un caso de Target Lekeage. Solo se utiliza en un modelo para demostrar como
    incluyendola en el entrenamiento, se puede obtener un muy buen puntaje.
    '''
    def __init__( self ):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        X['Sales_Contract_No'] = X['Sales_Contract_No'].replace('None', 0).astype(int)
        return X
```

```python
class ProductCategoryBTransformer( BaseEstimator, TransformerMixin ):
    '''
    Dejamos esta variable simplemente como 1 (si tiene alguna category B) o 0 (si el valor es 'None')
    '''
    def __init__( self):
        pass
      
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        X['Product_Category_B'] = X['Product_Category_B'] != 'None'
        X['Product_Category_B'] = X['Product_Category_B'].map({False:0, True:1})
        return X
```

```python
class Encoder( BaseEstimator, TransformerMixin ):
    def __init__( self):
        self.region_encoder = TargetEncoder()
        self.bureaucratic_code_encoder = TargetEncoder()
        self.account_owner_encoder = TargetEncoder()
        self.account_type_encoder = TargetEncoder()
        self.opportunity_type_encoder = TargetEncoder()
        self.product_name_encoder = TargetEncoder()
        self.last_modified_by_encoder = TargetEncoder()
        self.opportunity_owner_encoder = TargetEncoder()
        pass
      
    def fit( self, X, y = None ):
        y = X['Stage']
        self.region_encoder.fit(X['Region'],y)
        self.bureaucratic_code_encoder.fit(X['Bureaucratic_Code'],y)
        self.account_owner_encoder.fit(X['Account_Owner'],y)
        self.account_type_encoder.fit(X['Account_Type'],y)
        self.opportunity_type_encoder.fit(X['Opportunity_Type'],y)
        self.product_name_encoder.fit(X['Product_Name'],y)
        self.last_modified_by_encoder.fit(X['Last_Modified_By'],y)
        self.opportunity_owner_encoder.fit(X['Opportunity_Owner'],y)
        return self 
    
    def transform( self, X, y = None ):
        X['Region'] = self.region_encoder.transform(X['Region'])
        X['Bureaucratic_Code'] = self.bureaucratic_code_encoder.transform(X['Bureaucratic_Code'])
        X['Account_Owner'] = self.account_owner_encoder.transform(X['Account_Owner'])
        X['Account_Type'] = self.account_type_encoder.transform(X['Account_Type'])
        X['Opportunity_Type'] = self.opportunity_type_encoder.transform(X['Opportunity_Type'])
        X['Product_Name'] = self.product_name_encoder.transform(X['Product_Name'])
        X['Last_Modified_By'] = self.last_modified_by_encoder.transform(X['Last_Modified_By'])
        X['Opportunity_Owner'] = self.opportunity_owner_encoder.transform(X['Opportunity_Owner'])
        return X
```

```python
train_df, test_df = train_test_split(df, test_size=0.2)
```

```python
drop_features= ['Opportunity_ID','Last_Activity','ASP_(converted)_Currency','Quote_Type','Delivery_Quarter',
                'Source ','Account_Created_Date','ASP','Last_Modified_Date','Actual_Delivery_Date', 'Sales_Contract_No','TRF',
                'Total_Amount_Currency','Month','Delivery_Year','Territory','Submitted_for_Approval', 'Brand',
                'Currency', 'Quote_Expiry_Date','Total_Taxable_Amount_Currency','Opportunity_Name','Billing_Country',
                'ASP_Currency', 'ASP_(converted)', 
                'Prod_Category_A','Month','Product_Family','Price','Size','ID','Planned_Delivery_End_Date','Account_Name','Delivery_Terms','Product_Type']

train_preprocessing_pipeline = Pipeline( steps = [('useless_dropper', UselessDropper()),
                                                 ('outliers_handler', OutliersHandler()),
                                                 ('stage_encoder', StageEncoder())])

preprocessing_pipeline = Pipeline( steps = [ ('columns_dropper', ColumnsDropper(drop_features)),
                                             ('taxable_amount', TotalTaxableAmountCalculator()),
                                             ('created_date_proc', CreatedDateProcessor()),
                                             ('planned_del_date_proc', PlannedDeliveryDateProcessor()),
                                             ('prod_cat_B', ProductCategoryBTransformer()),
                                             ('target_encoder',Encoder())
                                            ] )

drop_features_lekeage = drop_features.copy()
drop_features_lekeage.remove('Sales_Contract_No')

lekeage_pipeline = Pipeline( steps = [ ('columns_dropper', ColumnsDropper(drop_features_lekeage)),
                                       ('taxable_amount', TotalTaxableAmountCalculator()),
                                       ('created_date_proc', CreatedDateProcessor()),
                                       ('planned_del_date_proc', PlannedDeliveryDateProcessor()),
                                       ('sales_contract', SalesContractTransformer()),
                                       ('prod_cat_B', ProductCategoryBTransformer()),
                                       ('target_encoder',Encoder())
                                    ] )
```
